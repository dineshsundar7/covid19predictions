```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#!pip install plotly
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
import datetime as dt
from datetime import timedelta
```

```python
datewise1=pd.read_csv(r'/content/sample_data/tnvaccinationcsv.csv')
datewise2=pd.read_csv(r'/content/sample_data/keralavaccinationcsv.csv')
datewise3=pd.read_csv(r'/content/sample_data/maharashtravaccinationcsv.csv')
datewise4=pd.read_csv(r'/content/sample_data/odishavaccinationcsv.csv')
datewise5=pd.read_csv(r'/content/sample_data/upvaccinationcsv.csv')
datewise6=pd.read_csv(r'/content/sample_data/apvaccinationcsv.csv')
datewise2.tail
```

```python
datewise1["date"]=pd.to_datetime(datewise1["date"])
datewise2["date"]=pd.to_datetime(datewise2["date"])
datewise3["date"]=pd.to_datetime(datewise3["date"])
datewise4["date"]=pd.to_datetime(datewise4["date"])
datewise5["date"]=pd.to_datetime(datewise5["date"])
datewise6["date"]=pd.to_datetime(datewise6["date"])
datewise2.head(30)
```

```python
datewise1=datewise1.groupby(["date"]).agg({"% of first dose":'sum'})
datewise1["Days Since"]=datewise1.index-datewise1.index.min()
datewise2=datewise2.groupby(["date"]).agg({"% of first dose":'sum'})
datewise2["Days Since"]=datewise2.index-datewise2.index.min()
datewise3=datewise3.groupby(["date"]).agg({"% of first dose":'sum'})
datewise3["Days Since"]=datewise3.index-datewise3.index.min()
datewise4=datewise4.groupby(["date"]).agg({"% of first dose":'sum'})
datewise4["Days Since"]=datewise4.index-datewise4.index.min()
datewise5=datewise5.groupby(["date"]).agg({"% of first dose":'sum'})
datewise5["Days Since"]=datewise5.index-datewise5.index.min()
datewise6=datewise6.groupby(["date"]).agg({"% of first dose":'sum'})
datewise6["Days Since"]=datewise6.index-datewise6.index.min()
```

```python
fig=px.bar(x=datewise1.index,y=datewise1["% of first dose"])
fig.update_layout(title="Total number of first dose vaccinations of Tamil nadu",
                  xaxis_title="Date",yaxis_title="Number of doses administered",)
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=datewise1.index, y=datewise1["% of first dose"],
                    mode='lines+markers',
                    name='tamil nadu vaccinations'))
fig.add_trace(go.Scatter(x=datewise2.index, y=datewise2["% of first dose"],
                    mode='lines+markers',
                    name='kerala vaccinations'))
fig.add_trace(go.Scatter(x=datewise3.index, y=datewise3["% of first dose"],
                    mode='lines+markers',
                    name='maharashtra vaccinations'))
fig.add_trace(go.Scatter(x=datewise4.index, y=datewise4["% of first dose"],
                    mode='lines+markers',
                    name='odisha vaccinations'))
fig.add_trace(go.Scatter(x=datewise5.index, y=datewise5["% of first dose"],
                    mode='lines+markers',
                    name='uttar pradhesh vaccinations'))
fig.add_trace(go.Scatter(x=datewise6.index, y=datewise6["% of first dose"],
                    mode='lines+markers',
                    name='andhra pradhesh vaccinations'))
fig.update_layout(title="Rate of vaccinations administered",
                 xaxis_title="Date",yaxis_title="Number of doses",legend=dict(x=0,y=1,traceorder="normal"))
fig.show()
```
