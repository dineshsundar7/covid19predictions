```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#!pip install plotly
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
import datetime as dt
from datetime import timedelta
```

```python
datewise1=pd.read_csv(r'/content/TAMILNADU-Copy.csv')
datewise1_vaccinations=pd.read_csv(r'/content/tnvaccinationcsv.csv')
datewise1_vaccinations.head(30)
```

```python
datewise1["date"]=pd.to_datetime(datewise1["date"])
datewise1
```

```python
datewise1_vaccinations["date"]=pd.to_datetime(datewise1_vaccinations["date"])
```

```python
datewise1=datewise1.groupby(["date"]).agg({"% of rec":'sum',"% of dead":'sum'})
datewise1["Days Since"]=datewise1.index-datewise1.index.min()
datewise1_vaccinations= datewise1_vaccinations.groupby(["date"]).agg({"% of first dose":'sum'})
datewise1_vaccinations["Days Since"]=datewise1_vaccinations.index-datewise1_vaccinations.index.min()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=datewise1.index, y=datewise1["% of rec"],
                    mode='lines+markers',
                    name='percentage of recovery'))
fig.add_trace(go.Scatter(x=datewise1_vaccinations.index, y=datewise1_vaccinations["% of first dose"],
                    mode='lines+markers',
                    name='percentage of vaccination'))
fig.update_layout(title="percentage of recovery and vaccination",
                 xaxis_title="Date",yaxis_title="%",legend=dict(x=0,y=1,traceorder="normal"))
fig.show()
```
